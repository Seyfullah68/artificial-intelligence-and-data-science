# scatter plot matrisi

```python
import seaborn as sns
iris=sns.load_dataset("iris")
df=iris.copy()
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df,hue="species");
```

```python
sns.pairplot(df,hue="species",markers=["o","s","D"]);
```

```python
sns.pairplot(df,kind="reg",hue="species");
```
